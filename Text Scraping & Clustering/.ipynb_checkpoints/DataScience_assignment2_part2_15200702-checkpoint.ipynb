```python
import glob
from os.path import expanduser
from sklearn.feature_extraction.text import CountVectorizer
from nltk.stem.porter import *
import nltk
```

```python
porter = PorterStemmer() #Using porter stemmer to output tokens
tokenize = CountVectorizer().build_tokenizer()
from sklearn.feature_extraction import text
stopwords = text.ENGLISH_STOP_WORDS
all_tokens=[]

# List of path all the article files
Files = (glob.glob("*.txt"))   
for f in Files:
    fd = open(str(f), "r")
    text = fd.read()
    tokens1 = tokenize(text.lower()) #Split the files into tokens
    newtokens = []
    for token in tokens1:
        if not token in stopwords and token.isalpha():
            #Removing the stop words and numerics
            token=porter.stem(token)
            newtokens.append(token)
    all_tokens.append(newtokens)
    #all_tokens is a list of preprocessed words of each text file as seperate list items
    fd.close()

    
```

```python
sentenced_tokens=[]
for word in all_tokens:
    sentence=""
    for j in word:
        sentence = sentence + j + " "#joining the tokens into sentences to get the term frequncy below
    sentenced_tokens.append(sentence)  #sentenced_token is a list made up of preprocessed "text" from the articles
```

```python
# totally 1131 items of article preprocessed text
print(len(sentenced_tokens))
```

```python
print(sentenced_tokens[0])
```

```python
#Using tfidf vectorizer to get sparse matrix of tfidf values
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer(lowercase = True,analyzer = 'word',
                                   use_idf=True, smooth_idf=True,min_df = 0.1,norm = 'l2') 
#tfidf_vectorizer using minimum values of term documents as 0.1 and cosine metric.
```

```python
x=tfidf_vectorizer.fit_transform(sentenced_tokens) #Vectorizer is fitted to form a spase matrix of tfidf values and stored in x as an object
print(x.toarray())
xArray = x.toarray()
```

```python
tfidf_vectorizer.vocabulary_ #Displays the term freqency of the words
```

```python
terms=tfidf_vectorizer.get_feature_names() #Displays the tokens in ascending order of name
print(terms)
```

```python
from sklearn import metrics
from sklearn.cluster import KMeans

silhouette = [] #List to store the silhouette values
sumofSquare = [] #List to store inertia values (ie; total within sum of squares)
for k in range(2,10):
    km = KMeans(n_clusters=k, init='random', max_iter=100, n_init=1) 
    #Kmeans using minimum 2 cluster and maximum 10. Random is chosen for random initialization during centriod formation
    km.fit(xArray)
    silhouette.append(metrics.silhouette_score(xArray, km.labels_, sample_size=1000)) #calculates the silhouette values
    sumofSquare.append(km.inertia_) #appending the silhouette values to the list

print("Silhouette values :\n")
print(silhouette)
print("\n\nWithin sum of Squares :\n")
print(sumofSquare)
```

```python
from matplotlib import pyplot #Using matplotlib to plot the silhouette and within sum of squares graph across each k 
pyplot.plot(silhouette)
pyplot.title("Silhoutte Graph")
pyplot.show()

pyplot.plot(sumofSquare)
pyplot.title("Within Sum of Square")
pyplot.show()

```

```python
#As we see from the above graph the The silhoutte is maximum for k = 2 and within sum of square are minimum for k = 2
#Between k = 2 and 3 there is not much difference in values for within sum of square for k = 2 is chosen as the optimum value of k
km = KMeans(n_clusters=2, init='random', max_iter=100, n_init=1)

km.fit(xArray[1:500])
clus = km.predict(xArray[500:])
print(clus)
```

```python
for k in range(0,4):
    words=[sentenced_tokens[y] for y in [i for i, y in enumerate(clus) if i == k]]
    print(words)#print the words in each cluster
```

```python
from sklearn.cluster import AgglomerativeClustering
#Hiearchial clustering to see what segregation of data can be achieved
model = AgglomerativeClustering(n_clusters=4,affinity="cosine",linkage="complete")
#Keeping the number of clusters as 2 and affinity as cosine
#Linkage is chosen as complete linkage as the metrix is cosine.

model.fit(xArray) #Fit the model
clustering = model.labels_.tolist()
print(clustering)
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
from matplotlib import pyplot as plt
plt.scatter(xArray[:, 0], xArray[:, 1], c=model.labels_,
                        cmap=plt.cm.spectral)
plt.title
plt.show()

# calculate full dendrogram
Z = linkage(xArray, 'ward')
plt.figure(figsize=(25, 10))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('sample index')
plt.ylabel('distance')
dendrogram(
    Z,
    leaf_rotation=90.,  # rotates the x axis labels
    leaf_font_size=8.,  # font size for the x axis labels
)
plt.show()
```
